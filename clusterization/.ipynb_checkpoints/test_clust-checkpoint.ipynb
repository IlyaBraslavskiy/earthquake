```python
import numpy as np
import numpy.linalg as la
import scipy.stats as st
import scipy.special as sp
import copy

class VBGMM:
    def __init__(self, X, m_0, b_0, inv_W_0, mu_0, alpha_0, m_k, b_k, W_k, mu_k, alpha_k, R_k):
        self.X = copy.deepcopy(X)
        self.m_0 = copy.deepcopy(m_0)
        self.b_0 = copy.deepcopy(b_0)
        self.inv_W_0 = copy.deepcopy(inv_W_0)
        self.mu_0 = copy.deepcopy(mu_0)
        self.alpha_0 = copy.deepcopy(alpha_0)

        self.m_k = copy.deepcopy(m_k)
        self.b_k = copy.deepcopy(b_k)
        self.W_k = copy.deepcopy(W_k)
        self.inv_W_k = np.linalg.inv(W_k)
        self.mu_k = copy.deepcopy(mu_k)
        self.alpha_k = copy.deepcopy(alpha_k) 
        self.R_k = copy.deepcopy(R_k)

    def UpdateParams(self):
        K = self.W_k.shape[0]
        N_k = np.sum(self.R_k, axis=0)
        mean_x_k = self.R_k.T.dot(self.X) / N_k[:,None]
        S_k = np.zeros_like(self.W_k)
        for k in range(K):
            x_cov = np.einsum("Bi, Bj -> Bij", self.X - x_mean[k,:], self.X - x_mean[k,:])
            S_k[k,:,:] = np.sum(self.R_k[:,k].reshape(K, 1, 1) * x_cov, axis=0) / N_k[k]

        self.alpha_k = self.alpha_0 + N_k
        self.b_k = self.b_0 + N_k

        self.m_k = (self.b_0 * self.m_0 + mean_x_k * N_k[:,None])

        self.inv_W_k = S_k * N_k.reshape(K, 1, 1) 
        cov_mean = np.einsum("Bi, Bj -> Bij", x_mean - self.m_0, x_mean - self.m_0)
        cov_mean *= self.b_0 * N_k / self.b_k 
        self.inv_W_k += cov_mean
        self.inv_W_k += self.inv_W_0
        self.W_k = la.inv(self.inv_W_k)
        self.mu_k = self.mu_0 + N_k + 1

    def UpdateR(self):
        D = self.X.shape[1]
        K = self.R_k.shape[1]

        for k in range(K):
            log_quadratic = np.einsum('...i,...i->...', (X - self.m_k[k,:]).dot(self.mu_k * self.W_k), (X - self.m_k[k,:]))
            log_quadratic += D / self.b_k[k]
            self.R_k[:,K] =  -0.5 * log_quadratic

        self.R_k += sp.digamma(self.alpha_k) - sp.digamma(np.sum(self.alpha_k))
        self.R_k += 0.5 * (D * np.log(2) + la.det(self.W_k) + sp.polygamma(D, self.mu_k * 0.5))
                           
        self.R_k -= np.max(self.R_k, axis=1)[:,None]
        self.R_k = np.exp(self.R_k)
        self.R_k /= np.sum(self.R_k, axis=1)[:,None]
    
    def ELBO(self):
        pass;

    def VariationalUpdates(self, max_iter):
        for i in range(max_iter):
            self.UpdateR()
            self.UpdateParams()
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=1000, centers=4, n_features=2)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0, init='random').fit(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_);
```

```python
m_0 = np.zeros(2)
mu_0 = 1.
b_0 = 1.
inv_W_0 = np.identity(2)
alpha_0 = np.array([0.01] * 10) 
m_k = X[np.random.randint(X.shape[0], size=10)]
b_k = np.array([1.] * 10)
W_k = np.repeat(np.identity(2)[None,:,:], 10, axis=0)
mu_k = np.array([1.] * 10)
alpha_k = np.array([0.01] * 10) 
R_k = np.repeat(np.array([1 / 10.] * 10)[None,:], X.shape[0], axis=0)
```

```python
gmm = VBGMM(X, m_0, b_0, inv_W_0, mu_0, alpha_0, m_k, b_k, W_k, mu_k, alpha_k, R_k)
```

```python
gmm.VariationalUpdates(1000)
```
